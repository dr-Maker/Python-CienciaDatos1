## Variable modalidad de estudio

```python

tabla_modalidad = data["Modalidad"]
tabla_modalidad


tabla_frecuencia_modalidad = data["Modalidad"].value_counts()
print(tabla_frecuencia_modalidad)


tabla_frecuencia_modalidad.plot(kind='bar', color=['black', 'red'])

# Personalizar el gráfico
plt.xlabel('JORNADA')
plt.ylabel('Frecuencia')
plt.title('ESTUDIANTES SEGUN JORNADA')

# Mostrar el gráfico
plt.show()

```

## Variable EDAD, histograma 

```python
data['Edad'].plot(kind='hist', bins=10, title='HISTOGRAMA EDAD')

# Personalizar el gráfico
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('HISTOGRAMA EDAD')

# Mostrar el gráfico
plt.show()

##Variable Semestre,

Tabla_frecuencia_semestre = data["Semestre"].value_counts()
#Tabla_frecuencia_semestre = Tabla_frecuencia_semestre.sort_values(by='Semestre')
print(Tabla_frecuencia_semestre)

Tabla_frecuencia_semestre.plot(kind='bar',x='Semestre', y='Frecuencia', color=['blue', 'black','red'])

# Personalizar el gráfico
plt.xlabel('ESTUDIANTES SEGUN SEMESTRE')
plt.ylabel('Frecuencia')
plt.title('SEMESTRE')

# Mostrar el gráfico
plt.show()


```

## Variable Abstraccion final, 


```python
tabla_abstraccion = data["Abstraccion final"]
print(tabla_abstraccion)

tabla_frecuencia_Abstraccion = data["Abstraccion final"].value_counts()
print(tabla_frecuencia_Abstraccion)


tabla_frecuencia_Abstraccion.plot(kind='bar', color=['blue', 'black','red'])

# Personalizar el gráfico
plt.xlabel('TIPO DE ABSTRACCION')
plt.ylabel('Frecuencia')
plt.title('ABSTRACCION FINAL')

# Mostrar el gráfico
plt.show()
```

## Matriz de correlaciones

```python

##Correlaciones
corr = pd.DataFrame({
    'ED': data['Edad'],
    'SE': data['Semestre'],
    'P1': data['Pregunta1'],
    'P2': data['Pregunta2']
})

print(corr)

correlacion = corr.corr().round(2)
print(correlacion)
```

```python

#Matriz de correlaciones

plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap=['#6D9EC1', 'white', '#E46726'], cbar=True, linewidths=0.5, linecolor='white')

# Personalizar el gráfico
plt.title('Correlacion de las Variables', size=18, ha='center')
plt.suptitle('Nivel de Respuesta de Encuesta', size=14, ha='center')
plt.show()
```
